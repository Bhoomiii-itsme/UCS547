# QUES 2
 Compare the following numpy arrays: 
 
arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[1, 2], [3, 4]])

```python
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[1,2],[3,4]])
```

```python
#Method 1
if np.array_equal(arr1,arr2):
    print('equal')
else:
    print('not equal')
```

```python
#Method 2
# This method first compares arrays element-wise using ==, then checks if all results are True with .all().
dif = arr1 == arr2
b = dif.all()
print(b)
```

# QUES 3
Find the most frequent value and their indice(s) in the following arrays:

i. x = np.array([1,2,3,4,5,1,2,1,1,1])

ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x=np.array([1,2,3,4,5,1,2,1,1,1])
y=np.array([1,1,1,2,3,4,2,4,3,3])
```

```python
#using bincount(restricted method,only int allowed,list can get very big)
xbincount=np.bincount(x)
print(xbincount)
ybincount=np.bincount(y)
print(ybincount)
xmax=np.argmax(xbincount)
print(xmax,np.where(xmax))
```

```python
#method 2
def most_freq_value(x=np.array([1,2,3,4,5,1,2,1,1,1])):
    num1,count1=np.unique(x,return_counts=True)
    freq1=zip(num1,count1)
    print("The unique numbers and their frequencies are: ", dict(freq1))
most_freq_value()
most_freq_value([1,1,1,2,3,4,2,4,3,3])
```
